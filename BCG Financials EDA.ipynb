#### Companies - Tesla, Inc., Apple Inc., Microsoft Corporation.
##### Import data
##### Calculate yearly trend for each metric

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# Import csv data
BCG_df = pd.read_csv('BCG_Financials.csv')
BCG_df.head()
```

```python
BCG_df.info()
```

```python
BCG_df.describe()
```

#### Total Revenue Growth rate
##### Apple
- Apple experienced a 2.8% reduction in its total revenue.
- However, in 2024 they experienced a 2.02% growth from the previous reduction.
##### Microsoft
- Microsoft has experienced revenue growth wit 6.8% in 2023 and 15% in 2024
##### Tesla
- In 2023, Tesla experienced 18.8% revenue growth.
- However, in 2024, the 0.9% revenue growth was small compared to the previous year's.

```python
# Calculate yearly trend for all metrics and add as a new column
BCG_df['Revenue_growth (%)'] = (BCG_df.groupby("Company")["Total Revenue ($M)"].pct_change() * 100).round(3)

BCG_df
```

#### Net Income Growth rate
##### Apple
- Apple experienced a 2.8% reduction in its net income in 2023.
- And a consecutive 3.3% reduction in 2024.
##### Microsoft
- Microsoft has experienced a 0.5% net income reduction in 2023.
- And a 21.8% net income increase in 2024 (The highest growth).
##### Tesla
- In 2023, Tesla experienced 18.9% (The highest growth)net income growth.
- However, in 2024 there was a huge reduction of its net income, 52.2% (The highest reduction overall).

```python
BCG_df['NetIncome_growth (%)'] = (BCG_df.groupby("Company")["Net Income ($M)"].pct_change() * 100).round(3)

BCG_df
```

#### Total Assets Growth rate
##### Apple
- Apple experienced a 0.04%% reduction in its total assets.
- However, in 2024 they experienced a 3.5% growth of assets.
##### Microsoft
- Microsoft has experienced asset growth with 12.9% in 2023 and 24.3% in 2024 (The highest).
##### Tesla
- Tesla has experienced asset growth with 29.5% (The highest)in 2023 and 14.5% in 2024.

```python
BCG_df['Assets_growth (%)'] = (BCG_df.groupby("Company")["Total Assets ($M)"].pct_change() * 100).round(3)

BCG_df
```

#### Total Liabilities Growth rate
##### Apple
- Apple experienced a 3.8% reduction in its total liabilities.
- However, in 2024 they experienced a 6.05% growth.
##### Microsoft
- Microsoft has experienced revenue growth with 3.7% in 2023 and 18.4% in 2024.
##### Tesla
- Tesla experienced an 18% liabilities increase in 2023 and 12.5% increase in 2024.

```python
BCG_df['Liabilities_growth (%)'] = (BCG_df.groupby("Company")["Total Liabilities ($M)"].pct_change() * 100).round(3)

BCG_df
```

#### Total Operations Cash flow Growth rate
##### Apple
- Apple experienced a 9.58% reduction in its total operations cash flow in 2023.
- However, in 2024 they experienced a 6.97% growth from the previous reduction.
##### Microsoft
- Microsoft has experienced a 1.6% reduction in its operations cash flow in 2023.
- While it experienced a 35.4% growth in 2024.
##### Tesla
- In 2023, Tesla experienced 9.9% reduction in operations cash flow.
- However in 2024, it experienced a 12.5% operations cash flow increase.

```python
BCG_df['Operations cashflow growth (%)'] = (BCG_df.groupby("Company")["Cash Flow from Operating Activities ($M)"].pct_change() * 100).round(3)

BCG_df
```

```python
# Change NaN values to be represented by 0
BCG_df.fillna(0, inplace = True)
```

#### Total revenue and net income over the 3 years
##### The highest earning company over the three years is Apple, Microsoft and Tesla in 3rd place.

```python
companies_summary = BCG_df.groupby("Company").agg({
    "Total Revenue ($M)": ["mean"],
    "Net Income ($M)": ["mean"],
}).round(3)

companies_summary
```

#### Net profit margin
##### A deep dive into the companies' net profit margin indicates that;
- Microsoft is the highest eaner with 36.6% in 2022 - 35.1% in 2024
- Followed by Apple with a range of 25.3% in 2022- 23.9% in 2024
- Teslas's was between 15.5%in 2022 - 7.3% in 2024

```python
BCG_df['Net profit margin %'] = ((BCG_df['Net Income ($M)'] / BCG_df['Total Revenue ($M)']) *100).round(3)

BCG_df
```
